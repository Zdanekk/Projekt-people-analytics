```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
url=r"C:\Users\xbots\Projekt-people-analytics\data\02_Interim\SelectedFeatures.csv"
#url="https://github.com/Zdanekk/Projekt-people-analytics/blob/main/data/02_Interim/SelectedFeatures.csv"
#people_data=pd.read_csv("C:\Users\xbots\Projekt-people-analytics\data\02_Interim")
people_data=pd.read_csv(url)
df = pd.DataFrame(people_data)
print(people_data.head(3))
```

```python
people_data['ID'] = people_data['ID'].astype(str)
# Wybieranie tylko kolumn z danymi liczbowymi
numeric_data = people_data.select_dtypes(include=['number'])

# Obliczanie średnich wartości dla kolumn z danymi liczbowymi
mean_values = numeric_data.mean()

# Wyświetlanie średnich wartości
print("Średnie wartości dla danych liczbowych:")
print(mean_values)

# Tworzenie wykresu
plt.figure(figsize=(10, 5))
plt.bar(mean_values.index, mean_values.values, color='blue')

# Dodanie tytułu i etykiet
plt.title('Średnie wartości danych liczbowych')
plt.xlabel('Kolumny')
plt.ylabel('Średnia')

# Ustawienie etykiet osi X na liczby 1, 2, 3, ...
plt.xticks(ticks=range(1, len(mean_values) + 1), labels=range(1, len(mean_values) + 1))

# Wyświetlanie wartości średnich na wykresie
for index, value in enumerate(mean_values.values):
    plt.text(index + 1, value, f'{value:.2f}', ha='center', va='bottom')
# Wyświetlanie wykresu
plt.show()
```

```python
# Liczenie wartości 0 w każdej kolumnie
zero_counts = (numeric_data == 0).sum()

# Znalezienie kolumny z największą liczbą wartości 0
max_zero_count = zero_counts.max()
columns_with_max_zeros = zero_counts[zero_counts == max_zero_count].index.tolist()

# Sortowanie kolumn według liczby wartości 0
sorted_zero_counts = zero_counts.sort_values(ascending=False)

# Wyświetlanie kolumn posortowanych według liczby wartości 0
print("\nKolumny posortowane według liczby wartości 0:")
print(sorted_zero_counts)


# Tworzenie wykresu
plt.figure(figsize=(12, 6))
plt.bar(range(1, len(sorted_zero_counts) + 1), sorted_zero_counts.values, color='orange')

# Dodanie tytułu i etykiet
plt.title('Liczba wartości 0 w każdej kolumnie')
plt.xlabel('Kolumny')
plt.ylabel('Liczba wartości 0')

# Ustawienie etykiet osi X na liczby 1, 2, 3, ...
plt.xticks(ticks=range(1, len(sorted_zero_counts) + 1), labels=range(1, len(sorted_zero_counts) + 1))

# Wyświetlanie liczby wartości 0 na słupkach wykresu
for index, value in enumerate(sorted_zero_counts.values):
    plt.text(index + 1, value, str(value), ha='center', va='bottom')

# Wyświetlanie wykresu
plt.show()
```

```python

# Liczenie wartości 4 w każdej kolumnie
four_counts = (numeric_data == 4).sum()

# Znalezienie kolumny z największą liczbą wartości 4
max_four_count = four_counts.max()
columns_with_max_fours = four_counts[four_counts == max_four_count].index.tolist()

# Sortowanie kolumn według liczby wartości 4
sorted_four_counts = four_counts.sort_values(ascending=False)

# Wyświetlanie kolumn posortowanych według liczby wartości 4
print("\nKolumny posortowane według liczby wartości 4:")
print(sorted_four_counts)

# Tworzenie wykresu
plt.figure(figsize=(12, 6))
plt.bar(range(1, len(sorted_four_counts) + 1), sorted_four_counts.values, color='orange')

# Dodanie tytułu i etykiet
plt.title('Liczba wartości 4 w każdej kolumnie')
plt.xlabel('Kolumny')
plt.ylabel('Liczba wartości 4')

# Ustawienie etykiet osi X na liczby 1, 2, 3, ...
plt.xticks(ticks=range(1, len(sorted_four_counts) + 1), labels=range(1, len(sorted_four_counts) + 1))

# Wyświetlanie liczby wartości 4 na słupkach wykresu
for index, value in enumerate(sorted_four_counts.values):
    plt.text(index + 1, value, str(value), ha='center', va='bottom')

# Wyświetlanie wykresu
plt.show()
```
